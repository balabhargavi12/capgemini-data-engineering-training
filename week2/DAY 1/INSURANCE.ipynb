```python
from pyspark.sql.functions import *

# Remove rows where customer_id is null (important key)
customers_clean = customers.dropna(subset=["customer_id"])

# Fill null values
customers_clean = customers_clean.fillna({
    "name": "Unknown",
    "age": 0,
    "city": "Unknown"
})

# Trim and standardize strings
customers_clean = customers_clean.withColumn("name", trim(col("name")))
customers_clean = customers_clean.withColumn("city", initcap(trim(col("city"))))

customers_clean.show()
```

```python
# Remove invalid premium values
policies_clean = policies.filter(col("premium") >= 0)

# Drop null keys
policies_clean = policies_clean.dropna(subset=["policy_id", "customer_id"])

# Fill null policy_type
policies_clean = policies_clean.fillna({
    "policy_type": "Unknown",
    "premium": 0
})

# Standardize strings
policies_clean = policies_clean.withColumn("policy_type", initcap(trim(col("policy_type"))))

# Ensure date format
policies_clean = policies_clean.withColumn("start_date", to_date(col("start_date")))

policies_clean.show()
```

```python
# Remove negative claim amounts
claims_clean = claims.filter(col("claim_amount") >= 0)

# Drop null keys
claims_clean = claims_clean.dropna(subset=["claim_id", "policy_id"])

# Fill null status
claims_clean = claims_clean.fillna({
    "status": "Unknown"
})

# Standardize status
claims_clean = claims_clean.withColumn("status", initcap(trim(col("status"))))

# Ensure date format
claims_clean = claims_clean.withColumn("claim_date", to_date(col("claim_date")))

claims_clean.show()
```

```python
agents_clean = agents.dropna(subset=["agent_id"])

agents_clean = agents_clean.fillna({
    "name": "Unknown",
    "region": "Unknown"
})

agents_clean = agents_clean.withColumn("name", trim(col("name")))
agents_clean = agents_clean.withColumn("region", initcap(trim(col("region"))))

agents_clean.show()
```

```python
customers_clean.createOrReplaceTempView("customers")
policies_clean.createOrReplaceTempView("policies")
claims_clean.createOrReplaceTempView("claims")
agents_clean.createOrReplaceTempView("agents")
policy_agent_clean.createOrReplaceTempView("policy_agent")
```

```python
%sql
WITH premium_cte AS (
    SELECT 
        customer_id,
        SUM(premium) AS total_premium
    FROM policies
    GROUP BY customer_id
),

claims_cte AS (
    SELECT 
        p.customer_id,
        SUM(c.claim_amount) AS total_claims
    FROM claims c
    JOIN policies p ON c.policy_id = p.policy_id
    GROUP BY p.customer_id
),

risk_cte AS (
    SELECT 
        c.customer_id,
        c.city,
        COALESCE(cl.total_claims, 0) AS total_claims,
        COALESCE(pr.total_premium, 0) AS total_premium,
        CASE 
            WHEN pr.total_premium = 0 THEN 0
            ELSE cl.total_claims / pr.total_premium
        END AS risk_score
    FROM customers c
    LEFT JOIN premium_cte pr ON c.customer_id = pr.customer_id
    LEFT JOIN claims_cte cl ON c.customer_id = cl.customer_id
),

rank_cte AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY city ORDER BY risk_score DESC) AS rank
    FROM risk_cte
)

SELECT *
FROM rank_cte
WHERE rank <= 3;
```

```python
%sql
WITH agent_premium AS (
    SELECT 
        pa.agent_id,
        a.name,
        a.region,
        SUM(p.premium) AS total_premium
    FROM policy_agent pa
    JOIN policies p ON pa.policy_id = p.policy_id
    JOIN agents a ON pa.agent_id = a.agent_id
    GROUP BY pa.agent_id, a.name, a.region
)

SELECT *,
       ROW_NUMBER() OVER (ORDER BY total_premium DESC) AS row_num,
       RANK() OVER (ORDER BY total_premium DESC) AS rank,
       DENSE_RANK() OVER (ORDER BY total_premium DESC) AS dense_rank
FROM agent_premium;
```
