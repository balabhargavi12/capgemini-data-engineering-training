**Replace NULL salary with 0**

```sql
select *,
coalesce(salary,0) as salary
from Employee630
```

**Replace NULL bonus with 1000**

```sql
select *,
coalesce(bonus,1000) as bonus
from employee630
```

**Show order amount, if NULL replace with 500**

```sql
select amount,
coalesce(amount,500) as Amount
from orders1
```

**Replace NULL stock with 0**

```sql
select *,
coalesce(stock,0) as Stock
from products
```

**Show employee earnings using:
 salary, if NULL use bonus
**

```sql
select *,
coalesce(salary,bonus) as Earnings
from employee630
```

**Show first available value:
 salary → bonus → 0
**

```sql
select *,
coalesce(salary,bonus,0) as AvailableValue
from employee630
```

**Show product price:
 price → 1000 (default)
**

```sql
select *,
coalesce(price,1000) as ProductPrice
from products
```

**Get customer payment:
 amount → discount → 0
**

```sql
select *,
coalesce(amount,discount,0) as Payment
from orders1
```
