```sql
-- 	Assign a unique row number to all employees based on salary (highest first). 
select *, row_number() over(order by salary desc) from employees_1 as row_numbers
```

```sql
--	Assign row numbers to employees within each department based on salary descending. 
select * , row_number() over(partition by department order by salary desc) as dept_rownum  from employees_1
```

```sql
--	Assign row numbers based on employee joining date (latest first). 
select *, row_number() over(order by join_date desc) as row_num from employees_1
```

```sql
-- 	Assign row numbers within each department based on earliest joining date. 
select * , row_number() over(partition by department order by join_date desc) as joindate_rownum  from employees_1
```

```sql
-- 5.	Assign row numbers to orders based on order date
select *, rank() over(order by salary desc) as sal_rank from employees_1
```

```sql
-- Assign dense rank to employees based on salary (highest first). 
select *, dense_rank() over(order by salary desc) as dense_rank from employees_1
```
